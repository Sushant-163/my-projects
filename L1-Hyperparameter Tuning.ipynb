```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

```python
df = load_digits()
```

```python
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(axes, df.images, df.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title("Training: %i" % label)
```

```python
n_samples = len(df.images)
data = df.images.reshape((n_samples, -1))
```

```python
data = data/16
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, df.target, test_size=0.3)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
baseline_rf = RandomForestClassifier()
```

```python
baseline_rf.fit(X_train, y_train)
```

```python
y_pred = baseline_rf.predict(X_test)
baseline_acc = accuracy_score(y_test, y_pred)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 20],
    'min_samples_split': [2, 5]
}
rf_grid = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf_grid,
    param_grid=param_grid,
    cv=3,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Best Parameters (Grid ):", grid_search.best_params_)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [100, 200],
    'max_depth': [None, 20],
    'min_samples_split': [2, 5]
}
rf_random = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(
    estimator=rf_random,
    param_distributions=param_dist,
    n_iter=4,  # since total combos = 2^3 = 8, we sample 4
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

print("Best Parameters (Random ):", random_search.best_params_)
```

```python
from sklearn.metrics import accuracy_score
# Grid Search model
grid_model = grid_search.best_estimator_
y_pred_grid = grid_model.predict(X_test)
grid_acc = accuracy_score(y_test, y_pred_grid)

# Random Search model
random_model = random_search.best_estimator_
y_pred_random = random_model.predict(X_test)
random_acc = accuracy_score(y_test, y_pred_random)

# Comparison output
print("\nModel Comparison:")
print(f"{'Model':<20}{'Accuracy'}")
print("-" * 30)
print(f"{'Baseline':<20}{baseline_acc:.4f}")
print(f"{'Grid Search':<20}{grid_acc:.4f}")
print(f"{'Random Search':<20}{random_acc:.4f}")
```
